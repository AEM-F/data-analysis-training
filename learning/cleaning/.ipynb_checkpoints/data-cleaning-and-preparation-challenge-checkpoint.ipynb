<h2>Dealing with outliers</h2>
<p>Often a dataset will have some points that are extreme outliers. Usually it is better to remove those few points, to have a more generalized model. When talking about outliers the first step we need to take is defining <b>range and limits</b> but also the <b>percentage of data</b>, those two are very <b>domain dependant</b>.</p>
<p>Some methodolgies that help in deciding the range and limits are:<ul><li>InterQuartile Range</li><li>Standard Deviation</li><li>Visualized or Domain Limit Value</li></ul>
</p>
The methodology I am going to apply is the Visualized or Domain Limit Value

<p>The label that I am thinking of predicting at the end of the challenges is the sales price. What I'm looking for is which feature is the most strongly correlated with the sales price. Next I can build scatterplots and analyse the visualtions for outliers.</p>

Looking at the scatterplot I can already see that there are a few points that are outliers. As quality goes up the price goes up but as we continue there are a few houses that buck this trend.<br>

<h2>Handling missing data</h2>

```python
df = pd.read_csv('data/ames_outliers_removed.csv')
```

In this section I'll start by taking a closer look at the dataset. Allready during the previous operations done on the dateset I noticed that the data had a lot of missing values

```python
df.info()
```

```python
df.sample(10)
```

The dataset seems to have a total of 2927 records(rows) and 82 features(columns).Apart from the NaN values heavly present in the dataset, I can see that something went wrong during the saving process. I have a new column "Unamed:0" that seems to have as values the indexes of the records. Next to it I can see that there is already an order column with the exact pourpose and a column "PID" that contains unique identifiers. The dataset can do without those duplications and the "PID" column as it does not have any value to a model.

```python
df = df.drop(['Unnamed: 0', 'Order', 'PID'], axis=1)
```

```python
df.head(5)
```

Now that those issues are addressed I want to look more into the missing data. I want to find out what is the total percentage of the missing data and what is the number and percentage for each feature.

```python
# get the number of missing data points per column
missingValuesCount = df.isnull().sum()
df.isnull().sum()
```

```python
totalCells = np.product(df.shape)
totalMissing = missingValuesCount.sum()

# total percent of data that is missing
(totalMissing/totalCells) * 100
```

```python
# custom method to return the percentages of the missing data
def percentMissing(dataFrame):
    missingValuesCount = df.isnull().sum()
    percentNaN = 100 * missingValuesCount / len(dataFrame)
    # sort and display only if there is missing data
    percentNaN = percentNaN[percentNaN > 0].sort_values()
    return percentNaN
```

```python
pNaN = percentMissing(df)
pNaN
```

```python
# let's visualize the missing percent series
sns.set(rc={'figure.figsize':(11.7,8.27)})
pBarPlot = sns.barplot(x=pNaN,y=pNaN.index)
# custom function to display the values next to bars
# from from https://medium.com/@dey.mallika/transform-your-graphs-with-seaborn-ea4fa8e606a6
initialx=0
for p in pBarPlot.patches:
    pBarPlot.text(p.get_width(),initialx+p.get_height()/8,"{:.2f}%".format(p.get_width()))
    initialx+=1
```

Looking at the percentage of data missing per feature column, I can see that most, and in some cases nearly all, of the rows from the last features have a null value. The features where there is an excessive amount of missing values can be dropped but before doing so, I will have to make a closer inspection. It may be a mistake in the way the data was filled in.<br> As I want to thoughtfully clean the dataset, before tackling those features I will start by addressing the smaller ones.

<h3>Fixing features with a low number of missing values</h3>

```python
pNaN[pNaN < 1]
```

```python
df[df['Electrical'].isnull()]
```

```python
# check if the row missing electrical is the same as the one missing from garage area
df[df['Electrical'].isnull()]['Garage Area']
```

```python
# check if the row missing electrical is the same as the one missing from garage cars
df[df['Electrical'].isnull()]['Garage Cars']
```

```python
# check if the row missing Garage Cars is the same as the one missing from Garage Area
df[df['Garage Cars'].isnull()]['Garage Area']
```

I can see that the row missing electrical is not the same one as the one missing the garage area. After the third attempt of checking if the record from one feature is encountered multiple times, I found out that Garage Cars and Garage Area share the same record. I can also see that all the features from the top till the fifth one(Garage cars) all have only one missing row. More often, when cleaning, it is advised to retain as much data as possible. Considering that the total number of rows is 2927, dropping a few rows that have the least percentage of missing data would not be a bad idea.

```python
df = df.dropna(subset=['Electrical','Garage Cars'])
```

```python
pNaN = percentMissing(df)
# lets visualize the percentage again and see if the change impacted other features as well
sns.set(rc={'figure.figsize':(11.7,8.27)})
pBarPlot = sns.barplot(x=pNaN,y=pNaN.index)
initialx=0
for p in pBarPlot.patches:
    pBarPlot.text(p.get_width(),initialx+p.get_height()/8,"{:.2f}%".format(p.get_width()))
    initialx+=1
```

From the previous visualization, we had 7 features that were missing one row. Now after the changes I can see that only 4 features remain. What is interesting is that most of the features that have a low number of missing values are related to the basement, so I will start by having a closer look at those features.

```python
df[df['Bsmt Half Bath'].isnull()]
```

```python
df[df['Bsmt Full Bath'].isnull()]
```

```python
df[df['Bsmt Unf SF'].isnull()]
```

Already, by exploring the missing values from just a few basement-related features I notice that rows 1341 and 1497 are present in both results. From the feature description, I can see that NA for the basement features signifies that there is no basement. As NaN indicates that there is no basement, for the numerical type basement features I will fill the NaN values with 0 and for the others that have a type string I will fill in the value none.

```python
# fill values for the basement numeric features
bsmtNumCols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF','Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[bsmtNumCols] = df[bsmtNumCols].fillna(0)
# fill values for the basement string features
bsmtStringCols = ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
df[bsmtStringCols] = df[bsmtStringCols].fillna('None')
```

```python
# lets have a look at the percentage of missing values again
pNaN = percentMissing(df)
pNaN
```

Great! From the results, it seems like I fixed all features that were missing under 0.7%. Apart from the masonry veneer type(MasVnrType) and the masonry veneer area(MasVnrArea), the remaining features have a high percentage of missing values. Before tackling those high percentage features I will make the same assumption, that the values are missing due to them not existing, for the masonry features and apply the same fixes.

```python
df['Mas Vnr Type'] = df['Mas Vnr Type'].fillna("None")
df['Mas Vnr Area'] = df['Mas Vnr Area'].fillna(0)
```

```python
# lets have a look at the percentage of missing values yet again
pNaN = percentMissing(df)
pNaN
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
pBarPlot = sns.barplot(x=pNaN,y=pNaN.index)
initialx=0
for p in pBarPlot.patches:
    pBarPlot.text(p.get_width(),initialx+p.get_height()/8,"{:.2f}%".format(p.get_width()))
    initialx+=1
```

<h3>Fixing features with a high number of missing values</h3>
<p>Due to the high number of missing values, it is recommended in this case to not focus on the rows but rather on the columns. If the majority of the information is missing I will remove these columns otherwise I will try to fill in those values.</p>

The garage-related features case is very similar to the one of the basement, meaning that if the garage type is NA that means that there is no garage. Most likely if there is no garage then the other garage features are simply missing because they do not exist.

```python
garageStringCols = ['Garage Type','Garage Finish','Garage Qual','Garage Cond']
df[garageStringCols] = df[garageStringCols].fillna('None')
df['Garage Yr Blt'] = df['Garage Yr Blt'].fillna(0)
```

```python
pNaN = percentMissing(df)
# lets visualize the percentage again to see the changes
sns.set(rc={'figure.figsize':(11.7,8.27)})
pBarPlot = sns.barplot(x=pNaN,y=pNaN.index)
initialx=0
for p in pBarPlot.patches:
    pBarPlot.text(p.get_width(),initialx+p.get_height()/8,"{:.2f}%".format(p.get_width()))
    initialx+=1
```

It seems like there are only 6 features with missing values remaining. Four of these features have a missing percentage of over 80%, so in this case, I believe it makes sense to just drop the entire columns.

```python
# droping the columns with a high percentage of missing values
df = df.drop(['Pool QC', 'Misc Feature', 'Alley', 'Fence'], axis = 1)
# visualize the changes
pNaN = percentMissing(df)
pNaN
```

With only two features remaining, I will have to figure out what to fill in those values, as not enough data is missing to just drop the rows or the columns.

```python
df['Fireplace Qu'] = df['Fireplace Qu'].fillna("None")
df['Fireplace Qu'].value_counts()
```

```python
df['Lot Frontage']
```

"LotFrontage: Linear feet of street connected to property"<br>
"Neighborhood: Physical locations within Ames city limits"<br>
Looking at the features description, I am going to make the domain knowledge assumption that the lot frontage is going to depend on the neighborhood. I am going to fill in the missing data with the average lot frontage for that neighborhood.

```python
df.groupby('Neighborhood')['Lot Frontage'].mean()
```

```python
df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(lambda val: val.fillna(val.mean()))
```

```python
df.isnull().sum()
```

```python
df['Lot Frontage'] = df['Lot Frontage'].fillna(0)
df.isnull().sum()
```

```python
df.to_csv("data/ames_no_missing_data.csv",index=False)
```

<h2>Dealing with categorical data</h2>
<p>Many machine learning models can not deal with categorical data set as strings. The solution is to convert these categories into "dummy" variables, otherwise known as "one-hot" encoding.<br>
In the feature description the first feature, MS SubClass, technically is a numerical feature because it is integer encoded but is more categorical. It implies some sort of linear relationship between all the values. The solution is to convert it to string type so that I can create dummy variables.
<p>

```python
df = pd.read_csv("data/ames_no_missing_data.csv")
df.head()
```

```python
df['MS SubClass'] = df['MS SubClass'].apply(str)
```

```python
# looks at the data type for every single column and checks if its an obj
dfObjs = df.select_dtypes(include='object')
dfObjs
```

```python
# looks at the data type for every single column and checks if it is not an obj
# in this case returing all the numeric type columns
dfNumeric = df.select_dtypes(exclude='object')
dfNumeric
```

Now that I have the data split by the types, what I am going to do is edit the string type features and then concatenate them with the numeric types. This will result in a lot of columns but later on, in the coming challenges, I will try to use regularization to check the importance of the features and remove the unwanted ones.

```python
dfObjDummies = pd.get_dummies(dfObjs, drop_first=True)
```

```python
big_df = pd.concat([dfNumeric,dfObjDummies], axis=1)
```

```python
big_df
```

```python
big_df.to_csv("data/ames_clean_data.csv",index=False)
```
